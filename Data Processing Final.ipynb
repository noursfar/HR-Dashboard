```python
def check_missing_values(df):
    """
    Vérifie s'il y a des cellules vides ou des valeurs manquantes (NA) dans chaque colonne de la dataframe.
    
    Args:
        df (pandas.DataFrame): La dataframe à vérifier.
        
    Returns:
        pandas.Series: Une série contenant le nombre de valeurs manquantes pour chaque colonne.
    """
    missing_values_count = df.isnull().sum()
    return missing_values_count
```

```python
def row_to_dictionary(row):
    """
    Crée un dictionnaire à partir d'une ligne de la dataframe où chaque valeur de 'MECM2', 'MECM3', etc.,
    est associée à la clé correspondante 'MECMn2', 'MECMn3', etc., si et seulement si chaque valeur est différente de NaN.
    
    Args:
        row (pandas.Series): Une ligne spécifique de la dataframe.
        
    Returns:
        dict: Un dictionnaire représentant les associations spécifiées pour la ligne donnée.
    """
    dictionary = {}
    for i in range(1, 7):
        mec_key = f"MEC{i}"
        mecn_key = f"MECn{i}"
        mec_value = row[mec_key]
        if pd.notnull(mec_value):
            dictionary[mec_value] = row[mecn_key]
    return dictionary
    
def row_to_column_value_dict(row, columns_to_extract):
    """
    Crée un dictionnaire pour une ligne de la dataframe où chaque clé est le nom de la colonne
    et chaque valeur est le contenu de la cellule de cette colonne dans la ligne donnée,
    à condition que le contenu de la cellule ne soit pas NULL (NaN).
    
    Args:
        row (pandas.Series): Une ligne spécifique de la dataframe.
        columns_to_extract (list): Liste des noms de colonnes pour lesquelles construire le dictionnaire.
        
    Returns:
        dict: Un dictionnaire représentant les associations spécifiées pour la ligne donnée.
    """
    dictionary = {}
    for column in columns_to_extract:
        cell_value = row[column]
        if pd.notnull(cell_value):
            dictionary[column] = cell_value
    return dictionary
    
def concatenate_dictionaries(dict1, dict2):
    """
    Concatène deux dictionnaires en utilisant la méthode update().
    
    Args:
        dict1 (dict): Premier dictionnaire.
        dict2 (dict): Deuxième dictionnaire.
        
    Returns:
        dict: Un nouveau dictionnaire contenant les clés et valeurs de dict1 et dict2.
    """
    new_dict = dict1.copy()  # Copie le premier dictionnaire pour conserver les données originales
    new_dict.update(dict2)
    return new_dict

def create_competences_metier_column(row):
    dict0 = row_to_dictionary(row)
    dict1 = row_to_column_value_dict(row, columns_to_extract)
    result_dict = concatenate_dictionaries(dict1, dict0)
    return result_dict
```

```python
columns_to_extract = ['TOrientation résultat', 'TCollaboration synergique', 'TEsprit entrepreneurial',
                      'TEmpathie Client', 'MAStratégie collaborative', 'MALeadership équipe',
                      'MAOrientation Client']
# Appliquer les opérations pour chaque ligne de la dataframe et créer la nouvelle colonne 'Competences_Metier'
df['Competences_Metier'] = df.apply(create_competences_metier_column, axis=1)
df['Competences_Metier'][0]
```

```python
dff=df[['Direction','Code_OTN','Code OMEA','Intitulé de Poste','Competences_Metier']]
dff1  = dff[dff['Competences_Metier'] != {}]
dff1
```

```python
# Créer une liste pour stocker les nouvelles lignes de la nouvelle dataframe
new_rows = []

# Itérer sur chaque ligne de la dataframe 'df'
for index, row in dff1.iterrows():
    intitule_poste = row['Intitulé de Poste']
    direction = row['Direction']
    Code_OTN=row['Code_OTN']
    CodeOMEA=row['Code OMEA']

    competences_metier_dict = row['Competences_Metier']
    
    # Itérer sur chaque compétence du dictionnaire
    for competence, niveau in competences_metier_dict.items():
        new_row = {
            'Direction':direction,
            'Code_OTN':Code_OTN,
            'Code OMEA':CodeOMEA,
            'Intitulé de Poste': intitule_poste,
            'Competence': competence,
            'Niveau': niveau
        }
        new_rows.append(new_row)

# Créer une nouvelle dataframe à partir de la liste de nouvelles lignes
new_df = pd.DataFrame(new_rows)

new_df
```

```python
def determine_type(competence):
    """
    Détermine le type en fonction de la compétence donnée.
    
    Args:
        competence (str): La compétence dont on souhaite déterminer le type.
        
    Returns:
        str: Le type de compétence ('Transversal', 'Managérial' ou 'Métier').
    """
    transversal_competences = ['TOrientation résultat', 'TCollaboration synergique',
                               'TEsprit entrepreneurial', 'TEmpathie Client']
    manageriel_competences = ['MAStratégie collaborative', 'MALeadership équipe',
                              'MAOrientation Client']
    
    if competence in transversal_competences:
        return 'Transversal'
    elif competence in manageriel_competences:
        return 'Managérial'
    else:
        return 'Métier'

# Appliquer la fonction determine_type pour créer la colonne 'Type' dans la nouvelle dataframe 'new_df'
new_df['Type'] = new_df['Competence'].apply(determine_type)

new_df
```
